# Water Quality Analysis using Python

I’ll start the water quality analysis task by importing the necessary Python libraries 

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

Inserting the DataSet

I can see null values in the first preview of this dataset itself, so before we go ahead, let’s remove all the rows that contain null values:

```python
data = data.dropna()
data.isnull().sum()
```

The PH column of this dataset is the column we need to predict because it contains values 0 and 1 that indicate whether the water is potable (1) or unfit (0) for consumption.

```python
plt.figure(figsize=(15,10))
sns.countplot(data.PH)
plt.title("Distribution of Unsafe and Safe Water")
plt.show()
```

let’s explore all the columns one by one. Let’s start by looking at the State column:

```python
import plotly.express as px
data = data
figure = px.histogram(data, x = "STATE", 
                      color = "PH", 
                      title= "Factors Affecting Water Quality: PH")
figure.show()
```

let’s explore all the columns one by one. Let’s start by looking at the Station Code column:

```python
figure = px.histogram(data, x = "STATION CODE", 
                      color = "PH", 
                      title= "Factors Affecting Water Quality: Hardness")
figure.show()
```

let’s explore all the columns one by one. Let’s start by looking at the Locations column:

```python
figure = px.histogram(data, x = "LOCATIONS", 
                      color = "PH", 
                      title= "Factors Affecting Water Quality: Solids")
figure.show()
```

let’s explore all the columns one by one. Let’s start by looking at the Temp column:

```python
figure = px.histogram(data, x = "Temp", 
                      color = "PH", 
                      title= "Factors Affecting Water Quality: Solids")
figure.show()
```
